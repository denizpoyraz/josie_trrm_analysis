```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python

def filter_solsonde(df):
    filtEN = df.ENSCI == 1
    filtSP = df.ENSCI == 0

    filtS10 = df.Sol == 1
    filtS05 = df.Sol == 0.5

    filtB10 = df.Buf == 1.0
    filtB05 = df.Buf == 0.5

    filterEN0505 = (filtEN & filtS05 & filtB05)
    filterEN1010 = (filtEN & filtS10 & filtB10)

    profEN0505 = df.loc[filterEN0505]
    profEN1010 = df.loc[filterEN1010]
    profEN0505_nodup = profEN0505.drop_duplicates(['Sim', 'Team'])
    profEN1010_nodup = profEN1010.drop_duplicates(['Sim', 'Team'])

    ###
    filterSP1010 = (filtSP & filtS10 & filtB10)
    filterSP0505 = (filtSP & filtS05 & filtB05)

    profSP1010 = df.loc[filterSP1010]
    profSP0505 = df.loc[filterSP0505]
    profSP1010_nodup = profSP1010.drop_duplicates(['Sim', 'Team'])
    profSP0505_nodup = profSP0505.drop_duplicates(['Sim', 'Team'])
    
    profEN0505_nodup = profEN0505_nodup.reset_index()
    profEN1010_nodup = profEN1010_nodup.reset_index()
    profSP0505_nodup = profSP0505_nodup.reset_index()
    profSP1010_nodup = profSP1010_nodup.reset_index()

    return profEN0505_nodup, profEN1010_nodup, profSP0505_nodup, profSP1010_nodup
```

```python
def make_beta_tables(dft, dftc):
    
    dft.loc['Median'] = dft.median()
    clist = list(dft)
    for i in clist:
#     err[j] = (np.nanquantile(dt[i].tolist(), 0.75) - np.nanquantile(dt[i].tolist(), 0.25)) / (2 * 0.6745)
        err = stats.median_abs_deviation(dftc.loc[0:, i].tolist())
#         print(i, err)
        dft.loc["MAD",i] = err
        
    return dft
    
```

```python
def make_beta_tables_mean(dft, dftc):
    
    dft.loc['Mean'] = dft.mean()
    clist = list(dft)
    for i in clist:
#     err[j] = (np.nanquantile(dt[i].tolist(), 0.75) - np.nanquantile(dt[i].tolist(), 0.25)) / (2 * 0.6745)
        err = np.nanstd(dftc.loc[0:, i].tolist())
#         print(i, err)
        dft.loc["std",i] = err
        
    return dft
```
